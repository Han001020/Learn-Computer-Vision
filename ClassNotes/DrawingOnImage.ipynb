```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Create Empty Image

```python
black_img = np.zeros(shape=(500,500,3),dtype=np.int16)
```

```python
plt.imshow(black_img)
```

## Rectangles

```python
cv2.rectangle(black_img,pt1=(380,10),pt2=(495,128),color=(0,255,15),thickness=5)
cv2.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=5)
plt.imshow(black_img)
```

## Circles

```python
cv2.circle(black_img,center=(100,100),radius=50,color=(180,120,200),thickness=5)
plt.imshow(black_img)
```

## Fill

```python
cv2.circle(black_img,center=(400,400),radius=40,color=(210,120,100),thickness=-1)
plt.imshow(black_img)
```

## Lines

```python
cv2.line(black_img,pt1=(0,0),pt2=(500,500),color=(255,100,0),thickness=5)
plt.imshow(black_img)
```
